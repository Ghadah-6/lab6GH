```python
df_missing = df.copy()

df_missing.loc[0:9, 'Price'] = np.nan

print("Missing values after introduction:")
print(df_missing.isna().sum())
```

```python
df_remove = df_missing.dropna()

print('df with NaN: ', df_missing.shape)
print('df without NaN: ', df_remove.shape)
```

```python
df_remove.isna().sum()
```

```python
df_mean = df_missing.copy()
mean_price = df_mean['Price'].mean()
df_mean['Price'] = df_mean['Price'].fillna(mean_price)

print(f"Mean Price used: ${mean_price:,.2f}")
df_mean.head(10)
```

```python
df_median = df_missing.copy()
median_price = df_median['Price'].median()
df_median['Price'] = df_median['Price'].fillna(median_price)

print(f"Median Price used: ${median_price:,.2f}")
df_median.head(10)
```

```python
df_median.isna().sum()
```

```python
plt.figure(figsize=(12, 8))

numerical_cols = ['Price', 'Avg. Area Income', 'Avg. Area House Age', 
                  'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms', 
                  'Area Population']

for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("IQR Statistics for Price:")
print(f"  Q1 (25th percentile): ${Q1:,.2f}")
print(f"  Q3 (75th percentile): ${Q3:,.2f}")
print(f"  IQR: ${IQR:,.2f}")
print(f"  Lower bound: ${lower:,.2f}")
print(f"  Upper bound: ${upper:,.2f}")

outliers = df[(df['Price'] < lower) | (df['Price'] > upper)]
print(f"\nNumber of outliers detected: {len(outliers)}")
print(f"Percentage of data: {len(outliers)/len(df)*100:.2f}%")
```

```python
print("Sample of outlier records (highest prices):")
outliers.nlargest(10, 'Price')[['Price', 'Avg. Area Income', 'Area Population']]
```

```python
df_no_outliers = df[(df['Price'] >= lower) & (df['Price'] <= upper)]

print('df with outliers: ', df.shape)
print('df without outliers: ', df_no_outliers.shape)
```

```python
lower_cap = df['Price'].quantile(0.05)  
upper_cap = df['Price'].quantile(0.95)  

df_capped = df.copy()
df_capped['Price'] = df_capped['Price'].clip(lower_cap, upper_cap)

print(f"Lower cap (5th percentile): ${lower_cap:,.2f}")
print(f"Upper cap (95th percentile): ${upper_cap:,.2f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].boxplot(df['Price'])
axes[0].set_title('Boxplot of Price (Original)')
axes[0].set_ylabel('Price ($)')

axes[1].boxplot(df_capped['Price'])
axes[1].set_title('Boxplot of Price (After Capping)')
axes[1].set_ylabel('Price ($)')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_minmax = df[['Price', 'Avg. Area Income', 'Area Population']].copy()

df_minmax[['Price', 'Avg. Area Income', 'Area Population']] = scaler.fit_transform(df_minmax)

print("Min-Max Normalization (Range: 0 to 1):")
df_minmax.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_zscore = df[['Price', 'Avg. Area Income', 'Area Population']].copy()

df_zscore[['Price', 'Avg. Area Income', 'Area Population']] = scaler.fit_transform(df_zscore)

print("Z-Score Standardization (Mean = 0, Std = 1):")
df_zscore.head()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df_zscore.corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap (Before PCA)")
plt.show()
```

```python
correlation = df['Price'].corr(df['Avg. Area Income'])
print(f"Correlation between Price and Avg. Area Income: {correlation:.4f}")

correlation2 = df['Avg. Area Number of Rooms'].corr(df['Avg. Area Number of Bedrooms'])
print(f"Correlation between Rooms and Bedrooms: {correlation2:.4f}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

all_numerical_features = ['Price', 'Avg. Area Income', 'Avg. Area House Age', 
                          'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms', 
                          'Area Population']

scaler = StandardScaler()
df_zscore = df.copy()  # Start with original dataframe
df_zscore[all_numerical_features] = scaler.fit_transform(df[all_numerical_features])

X = df_zscore[all_numerical_features]

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print(f"Total variance explained by 2 components: {pca.explained_variance_ratio_.sum()*100:.2f}%")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(principal_components[:,0], principal_components[:,1], alpha=0.5)
plt.title("PCA Projection (2 Components)")
plt.xlabel(f"Principal Component 1 ({pca.explained_variance_ratio_[0]*100:.1f}%)")
plt.ylabel(f"Principal Component 2 ({pca.explained_variance_ratio_[1]*100:.1f}%)")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
pca_full = PCA()
pca_full.fit(X)

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(pca_full.explained_variance_ratio_)+1), 
         pca_full.explained_variance_ratio_.cumsum(), 'bo-')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA - Cumulative Explained Variance')
plt.grid(True, alpha=0.3)
plt.show()
```
